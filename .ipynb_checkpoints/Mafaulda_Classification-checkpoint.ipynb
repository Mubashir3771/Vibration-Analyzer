# Building a Web app

```python
model = load_model('best_model')
```

```python
def predict(model, input_df):
    predictions_df = predict_model(estimator = model, data = input_df)
    predictions = predictions_df['Label'][0]
    return predictions
```

```python
def run():

    from PIL import Image
    image = Image.open("D:\pycaret classification using full data\images\logo.png")
    image_mafaulda = Image.open("D:\pycaret classification using full data\images\Mafaulda.png")

    st.image(image,use_column_width=False)

    add_selectbox = st.sidebar.selectbox(
    "How would you like to predict?",
    ("Online", "Batch"))

    st.sidebar.info('This app is created to predict machinery faults')
    
    st.sidebar.image(image_mafaulda)
    st.sidebar.success('https://www02.smt.ufrj.br/~offshore/mfs/page_01.html#SEC1')

    st.title("Machinery Faults Prediction App")

    if add_selectbox == 'Online':

        tachometer_signals = st.number_input('Tachometer', min_value=-1, max_value=1, value=0)
        uba_axial_direction = st.number_input('Underhang Bearing Accelerometer Axial Direction', min_value=-1, max_value=1, value=0)
        uba_radiale_direction = st.number_input('Underhang Bearing Accelerometer Radiale Direction', min_value=-1, max_value=1, value=0)
        uba_tangential_direction = st.number_input('Underhang Bearing Accelerometer Tangential Direction', min_value=-1, max_value=1, value=0)
        oba_axial_direction = st.number_input('Overhang Bearing Accelerometer Axial Direction', min_value=-1, max_value=1, value=0)
        oba_radiale_direction = st.number_input('Overhang Bearing Accelerometer Radiale Direction', min_value=-1, max_value=1, value=0)
        oba_tangential_direction = st.number_input('Overhang Bearing Accelerometer Tangential Direction', min_value=-1, max_value=1, value=0)
        microphone = st.number_input('Microphone', min_value=-1, max_value=1, value=0)
        
        output=""

        input_dict = {'tachometer_signals' : tachometer_signals, 'uba_axial_direction' : uba_axial_direction, 'uba_radiale_direction' : uba_radiale_direction, 'uba_tangential_direction' : uba_tangential_direction, 'oba_axial_direction' : oba_axial_direction, 'oba_radiale_direction' : oba_radiale_direction, 'oba_tangential_direction' : oba_tangential_direction, 'microphone' : microphone}
        input_df = pd.DataFrame([input_dict])

        if st.button("Predict"):
            output = predict(model=model, input_df=input_df)
            output = '$' + str(output)

        st.success('The output is {}'.format(output))

    if add_selectbox == 'Batch':

        file_upload = st.file_uploader("Upload csv file for predictions", type=["csv"])

        if file_upload is not None:
            data = pd.read_csv(file_upload)
            predictions = predict_model(estimator=model,data=data)
            st.write(predictions)
            
if __name__ == '__main__':
    run()
```

```python
streamlit run.py
```
